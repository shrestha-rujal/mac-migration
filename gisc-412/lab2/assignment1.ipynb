```python
import geopandas as gpd
from shapely.geometry import MultiPolygon, Polygon

# Example MultiPolygons representing 2 or more subdivided blocks
multi1 = MultiPolygon([
    Polygon([(174.77000, -36.86000), (174.77500, -36.86000), (174.77500, -36.85500), (174.77000, -36.85500)]),
    Polygon([(174.77600, -36.86050), (174.77900, -36.86050), (174.77900, -36.85800), (174.77600, -36.85800)])
])

multi2 = MultiPolygon([
    Polygon([(174.78500, -36.86500), (174.79000, -36.86500), (174.79000, -36.86000), (174.78500, -36.86000)])
])

multi3 = MultiPolygon([
    Polygon([(174.80000, -36.87000), (174.80500, -36.87000), (174.80500, -36.86500), (174.80000, -36.86500)]),
    Polygon([(174.80600, -36.87050), (174.80900, -36.87050), (174.80900, -36.86700), (174.80600, -36.86700)])
])

# Create the GeoDataFrame
land_parcels_gdf = gpd.GeoDataFrame({
    'Block_ID': [1, 2, 3],
    'Land_Use_Type': ['Residential', 'Commercial', 'Residential'],
    'geometry': [multi1, multi2, multi3]
}, crs='EPSG:4326')

# Preview
land_parcels_gdf
```

```python
exploded = land_parcels_gdf.explode()

exploded
```

```python
exploded['area'] = exploded.area
```

```python
data = {
    'Animal_ID': [1, 2, 3, 4, 5],
    'timestamp': [
        '21/07/2023 10:00',
        '21/07/2023 10:15',
        '21/07/2023 11:30',
        '21/07/2023 11:45',
        '21/07/2023 12:00'
    ],
    'geometry': [
        Point(174.77000, -36.86000),
        Point(174.77500, -36.86000),
        Point(174.78000, -36.86200),
        Point(174.78500, -36.86200),
        Point(174.79000, -36.86000)
    ]
}

# Create GeoDataFrame and parse timestamps
animal_locations_gdf = gpd.GeoDataFrame(data, crs='EPSG:4326')

animal_locations_gdf['timestamp'] = pd.to_datetime(animal_locations_gdf['timestamp'])

animal_locations_gdf
```

```python
def categorize_time(hour):
  if (hour >= 6) and (hour < 12): return 'Morning'
  elif (hour >= 12) and (hour < 17): return 'Afternoon'
  else: return 'Night'
```

```python
animal_locations_gdf['time_category'] = animal_locations_gdf.timestamp.dt.hour.apply(categorize_time)
```

```python
import pandas as pd

# Sample data
data = {
    'Property_ID': [1, 2, 3, 4, 5, 6],
    'District': ['District_A', 'District_A', 'District_B', 'District_B', 'District_B', 'District_C'],
    'Business_Type': ['Restaurant', 'Retail', 'Office', 'Restaurant', 'Retail', 'Entertainment'],
    'Latitude': [-36.848, -36.846, -36.855, -36.853, -36.852, -36.861],
    'Longitude': [174.778, 174.781, 174.793, 174.790, 174.789, 174.795]
}

# Create DataFrame
commercial_properties_df = pd.DataFrame(data)

# Preview
commercial_properties_df
```

```python
comm_gpd = gpd.GeoDataFrame(commercial_properties_df, geometry=gpd.points_from_xy(commercial_properties_df['Longitude'], commercial_properties_df['Latitude'], crs='EPSG:4326'))

comm_gpd
```

```python
comm_gpd.to_crs(epsg=2193, inplace=True)

comm_gpd
```

```python
grouped_by_district = gpd.GeoDataFrame(comm_gpd.groupby('District', as_index=False)['geometry'].apply(lambda x: MultiPoint(x.tolist())), geometry='geometry')

grouped_by_district
```

```python
grouped_by_district['centroid'] = grouped_by_district.centroid
```
